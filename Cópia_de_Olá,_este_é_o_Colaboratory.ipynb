```
import pandas as pd


pop = pd.read_csv('world_population.csv')
pop.head()
print(pop["Continent"].unique())

pop_str = pop.set_index(["Continent","Country/Territory"]).sort_index()
print(pop_str)

b_countries = pop_str.loc[("South America","Argentina"):("South America","Brazil")]
print(b_countries)


teste = pop_str.xs("South America",level = "Continent")

sumary_south_america = pop[pop["Continent"] == "South America"]["World Population Percentage"].sum()
sumary_africa = pop[pop["Continent"] == "Africa"]["World Population Percentage"].sum()
sumary_Asia = pop[pop["Continent"] == "Asia"]["World Population Percentage"].sum()
sumary_Europe = pop[pop["Continent"] == "Europe"]["World Population Percentage"].sum()
sumary_north_america = pop[pop["Continent"] == "North America"]["World Population Percentage"].sum()
sumary_oceania = pop[pop["Continent"] == "Oceania"]["World Population Percentage"].sum()

total = sumary_south_america+sumary_africa+sumary_Asia+sumary_Europe+sumary_oceania+sumary_north_america
print(total)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


pop = pd.read_csv('world_population.csv')
print(pop["Continent"].unique())
#['Asia' 'Europe' 'Africa' 'Oceania' 'North America' 'South America']
sumary_asia = pop[pop["Continent"] == "Asia"]["World Population Percentage"].sum()
sumary_europe = pop[pop["Continent"] == "Europe"]["World Population Percentage"].sum()
sumary_africa = pop[pop["Continent"] == "Africa"]["World Population Percentage"].sum()
sumary_oceania = pop[pop["Continent"] == "Oceania"]["World Population Percentage"].sum()
sumary_north_america = pop[pop["Continent"] == "North America"]["World Population Percentage"].sum()
sumary_south_america = pop[pop["Continent"] == "South America"]["World Population Percentage"].sum()

total = sumary_asia+sumary_africa+sumary_europe+sumary_oceania+sumary_north_america+sumary_north_america

cont_perc_dict = {
    "Continent":["Asia","Europe","Africa","Oceania","North America","South America"],
    "World Population Percentage":[sumary_asia,sumary_europe,sumary_africa,sumary_oceania,sumary_north_america,sumary_south_america]
}

sns.barplot(data = cont_perc_dict,x = "Continent",y="World Population Percentage",width=0.5)
plt.xticks(rotation=45)
plt.show()
```

```
#['Asia' 'Europe' 'Africa' 'Oceania' 'North America' 'South America']
Asia_Density = pop[pop["Continent"] == "Asia"]["Density (per km²)"].sum()
Europe_Density = pop[pop["Continent"] == "Europe"]["Density (per km²)"].sum()
Africa_Density = pop[pop["Continent"] == "Africa"]["Density (per km²)"].sum()
Oceania_Density = pop[pop["Continent"] == "Oceania"]["Density (per km²)"].sum()
North_America_Density = pop[pop["Continent"] == "North America"]["Density (per km²)"].sum()
South_America_Density = pop[pop["Continent"] == "South America"]["Density (per km²)"].sum()

total = Asia_Density+Europe_Density+Africa_Density+Oceania_Density+North_America_Density+South_America_Density

Densidade_Total = {
    "Continent":["Asia","Europe","Africa","Oceania","North America","South America"],
    "Density (per km²)":[Asia_Density,Europe_Density,Africa_Density,Oceania_Density,North_America_Density,South_America_Density]
}

sns.barplot(data = Densidade_Total,x = "Continent",y="Density (per km²)",width=0.5)
plt.xticks(rotation=45)
plt.show()
print(Densidade_Total)
```

```
sumary = pop.groupby("Continent").agg(
    Total_2022 = ("2022 Population","sum"),
    Total_2020 = ("2020 Population","sum")
)
sumary["Crescimento_Entre_2020_2022"] = sumary['Total_2022'] - sumary['Total_2020']
print(sumary)

sns.barplot(data = sumary,x="Continent",y="Crescimento_Entre_2020_2022",width=0.5)
plt.xticks(rotation=45)
plt.ticklabel_format(style="plain",axis="y")
min_y = min(sumary["Crescimento_Entre_2020_2022"]) * 1.2
plt.ylim(min_y, max(sumary["Crescimento_Entre_2020_2022"]) * 1.1)
plt.axhline(0, color='black', linestyle='dashed')
plt.show()
```
